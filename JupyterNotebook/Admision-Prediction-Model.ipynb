```python
scores = []
for i in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=i)
    lr = LinearRegression()
    lr.fit(X_train, y_train)
    y_predict = lr.predict(X_test)
    scores.append(r2_score(y_test, y_predict))
```

```python
sorted(list(enumerate(scores)),reverse=True, key=lambda x :x[1])
```

```python
# Let's create a function to create adjusted R-Squared
def adj_r2(x,y):
    r2 = lr.score(x,y)
    n = x.shape[0]
    p = x.shape[1]
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=637)
lr = LinearRegression()
lr.fit(X_train, y_train)
y_predict = lr.predict(X_test)
print(r2_score(y_test, y_predict))
```

r2_score(y_test, y_predict) is used to evaluate the model's performance on unseen data (test set), while LinearRegression.score(X_train, y_train) is used to evaluate the model's performance on the training data itself. 

```python
print(lr.score((X_train), pd.DataFrame(y_train)))
print(adj_r2(pd.DataFrame(X_train), pd.DataFrame(y_train)))
```

```python
print(lr.score((X_test), pd.DataFrame(y_test)))
print(adj_r2(pd.DataFrame(X_test), pd.DataFrame(y_test)))
```

__So it looks like our model r2 score is less on the test data.__

Let's see if our model is overfitting our training data.

```python
from sklearn.linear_model  import Ridge,Lasso,RidgeCV, LassoCV, ElasticNet, ElasticNetCV
```

```python
# Lasso Regularization
# LassoCV will return best alpha and coefficients after performing 10 cross validations
lasscv = LassoCV(alphas = None,cv =10, max_iter = 100000)
lasscv.fit(X_train, y_train)
```

```python
# best alpha parameter
alpha = lasscv.alpha_
alpha
```

```python
#now that we have best parameter, let's use Lasso regression and see how well our data has fitted before

lasso_reg = Lasso(alpha)
lasso_reg.fit(X_train, y_train)
```

```python
lasso_reg.score(X_test, y_test)
```

Our r2_score for test data (89.26%) comes same as before using regularization. So, it is fair to say our OLS model did not overfit the data.


# PICKLE FILE

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, np.log(y), test_size=0.25, random_state=637)

step1 = StandardScaler()

step2 = LinearRegression()

pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

pipe.fit(X_train, y_train)
y_predict = pipe.predict(X_test)
print(r2_score(y_test,y_predict))
```
